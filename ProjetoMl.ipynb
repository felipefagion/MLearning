###  Vari√°veis que vamos trabalhar



- <p><strong>Age</strong>: Idade do cliente.</p>

- <p><strong>Job</strong>: Tipo de trabalho do cliente.</p>

- <p><strong>Marital</strong>: Estado civil do cliente.</p>

- <p><strong>Education</strong>: N√≠vel de educa√ß√£o do cliente.</p>

- <p><strong>Default</strong>: Indicador se o cliente tem cr√©dito em inadimplente.</p>

- <p><strong>Balance</strong>: Saldo m√©dio anual do cliente em euros.</p>

- <p><strong>Housing</strong>: Indicador se o cliente tem empr√©stimo habita√ß√£o.</p>

- <p><strong>Previous</strong>: N√∫mero de contatos realizados antes desta campanha e para este cliente.</p>

- <p><strong>Poutcome</strong>: Resultado da campanha de marketing anterior.</p>

```python
df.drop(['marital', 'education', 'contact'], axis=1, inplace=True)
```

```python
# Importando bibliotecas necess√°rias
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Criando uma figura para o gr√°fico com tamanho espec√≠fico
fig = plt.figure(figsize=(12,8))

# Transformando a vari√°vel alvo 'depositou' usando LabelEncoder para torn√°-la num√©rica
df["depositou"] = LabelEncoder().fit_transform(df["depositou"])

# Selecionando apenas as colunas num√©ricas do DataFrame
df_numerico = df.select_dtypes(exclude="object")

# Calculando a matriz de correla√ß√£o entre as vari√°veis num√©ricas
corr_numeric = df_numerico.corr()

# Plotando um mapa de calor da matriz de correla√ß√£o
sns.heatmap(corr_numeric, cbar=True)
plt.title("Matriz de Correla√ß√£o", fontsize=16)
plt.show()
```

```python
df_numerico.head()
```

```python
clf = DecisionTreeClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_numerico.drop("depositou",axis=1),df_numerico["depositou"],test_size=0.3)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
clf=clf.fit(X_train,y_train)
clf.feature_importances_
for feature, importancia in zip (df_numerico.columns, clf.feature_importances_):
    print(" {}:{}".format(feature, importancia))
```

<div align="center">
 <h1> Avalia√ß√£o do Modelo </h1>
 </div>

```python
#M√©dia da precis√£o
resultado= clf.predict(X_test)
clf.score(X_test, y_test)
```

```python
#Precis√£o do modelo
print(metrics.classification_report(y_test, resultado))
```

```python
#Importando bibliotecas para a cria√ß√£o do painel
from ipywidgets import interactive
from IPython.display import SVG,display
from graphviz import Source
from sklearn.tree import export_graphviz
```

```python
# Definindo as vari√°veis de caracter√≠sticas (X) e a vari√°vel alvo (y).
X, y = df_numerico.drop("depositou", axis=1), df_numerico["depositou"]

# Rotulando as caracter√≠sticas para uso posterior na visualiza√ß√£o da √°rvore de decis√£o.
features_label = df_numerico.drop("depositou", axis=1).columns

# Rotulando as classes para uso posterior na visualiza√ß√£o da √°rvore de decis√£o.
class_label = ["0", "1"]

# Definindo uma fun√ß√£o para plotar uma √°rvore de decis√£o com base nos par√¢metros fornecidos.
def plot_tree(crit, split, depth, min_samples_split, min_samples_leaf=0.2):

    # Criando um estimador de √°rvore de decis√£o com os par√¢metros fornecidos.
    estimador = DecisionTreeClassifier(
        random_state=0,
        criterion=crit,
        splitter=split,
        max_depth=depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf
    )

    # Treinando o modelo com os dados.
    estimador.fit(X, y)

    # Criando um gr√°fico da √°rvore de decis√£o usando Graphviz.
    graph = Source(export_graphviz(estimador,
                                   out_file=None,
                                   feature_names=features_label,
                                   class_names=class_label,
                                   impurity=True,
                                   filled=True))
    # Exibindo a √°rvore de decis√£o.
    display(SVG(graph.pipe(format="svg")))

    # Retornando o modelo treinado para uso posterior.
    return estimador

# Criando uma interface interativa para ajustar os par√¢metros e visualizar diferentes √°rvores de decis√£o.
inter = interactive(plot_tree,
                    crit=["gini", "entropy"],
                    split=["best", "random"],
                    depth=[1, 2, 3, 4, 5, 20, 30],
                    min_samples_split=(1, 5),
                    min_samples_leaf=(1, 5))

# Exibindo a interface interativa.
display(inter)
```
